```python
import cv2
import math
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread('data/checkerboard_clean.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
src = gray.astype(np.float32)
dst = cv2.cornerHarris(src, blockSize=2, ksize=11, k=0.04)
plt.imshow(dst)
```

```python
rows, cols = dst.shape
th = 0.9 * np.max(dst)
result = np.copy(img)

for r in range(rows):
    for c in range(cols):
        if dst[r, c] > th:
            result = cv2.circle(result, (c,r), 5, (255,0,0), -1)
            
plt.imshow(result)
```

```python
img = cv2.imread('data/checkerboard.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

src = gray.astype(np.float32)
dst = cv2.cornerHarris(src, blockSize=2, ksize=3, k=0.04)
plt.subplot(121), plt.imshow(img)
plt.subplot(122), plt.imshow(dst)
```

```python
rows, cols = dst.shape
th = 0.2 * np.max(dst)
result = np.copy(img)

for r in range(rows):
    for c in range(cols):
        if dst[r, c] > th:
            result = cv2.circle(result, (c,r), 1, (255,0,0), -1)
            
plt.imshow(result)
```
