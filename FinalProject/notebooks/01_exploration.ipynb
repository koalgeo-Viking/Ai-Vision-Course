```python
# Helper function for displaying images
def show_image(img, title='Image', cmap=None):
    plt.figure(figsize=(8, 8))
    if cmap:
        plt.imshow(img, cmap=cmap)
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```
